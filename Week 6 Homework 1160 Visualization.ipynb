```python
df=pd.read_csv("diabetes.data",sep="\s+")
df.head()
```

```python
sns.relplot(x='BMI',y='BP',hue ='AGE',row='SEX',data=df)
```

```python
sns.relplot(x='AGE',y='BMI',hue ='BP',row='S4',col="SEX",data=df)
```

```python
#BoxPlot
```

```python
import seaborn as sns
sns.catplot(x="SEX", y= "BP",hue='SEX', kind='box',data=df)
```

```python
# by looking at this plot we can analyize BP is higher in male as compare to female 
```

```python
sns.catplot(x = "SEX", y = "BP", kind='bar',data=df)
```

```python
sns.catplot(x = "BP", kind='count',palette='ch:0.45',data=df)
```

```python
#Scatter plots
for col1_idx, column1 in enumerate(df.columns):
    for col2_idx, column2 in enumerate(df.columns):
        if col1_idx < col2_idx:
            fig, axes = plt.subplots(1, 1, figsize=(5, 5))
            axes.scatter(df[column1], df[column2], label=f'{column1} to {column2}', color='blue', marker='x')
            axes.set_title(f'{column1} to {column2}')
            axes.set_xlabel(column1)
            axes.set_ylabel(column2)
            axes.legend()
        
```
